### Visualization of the measured Gflop/s vs. N

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data

```python
data = pd.read_csv("../data/results_esteban_dev-amd20.csv")
data.head()
```

Visualize plot

```python
plt.plot(data['matrix size'], data['matrix multiplication time'])
plt.axhline(y= 0.010, color='r', linestyle='--')  # horizontal line
plt.annotate("theoretical peak = ###", (0, 0.011))  # text with xy-coord
plt.xlabel('N', size=14)
plt.ylabel('Gflop/s', size=14)
plt.title("Measured Gflop/s vs. N on 'dev-amd20'", size=14)
plt.grid(alpha=.2)
plt.savefig("sample.png", dpi=300)  # save plot with high resolution
```
